```python
from Perceptron import *
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# values
t = [1,1,0,0]
f = [1,0,1,0]

# and, or, xor, nand
plt.subplot(2, 2, 1)
plt.title('Logic "And" is Linearly Separable')
plt.scatter(t[1:], f[1:], color='red')
plt.scatter(t[0],f[0], color='blue')
plt.plot((1.1,.1), 'g--')

plt.subplot(2, 2, 2)
plt.title('Logic "Or" is Linearly Separable')
plt.scatter([t[0:3]], [f[0:3]], color='red')
plt.scatter(t[3],f[3], color='blue')
plt.plot((.9,-.1), 'g--')

plt.subplot(2, 2, 3)
plt.title('Logic "XOr" is Not Linearly Separable')
plt.scatter([t[0],t[3]], [f[0],f[3]], color='red')
plt.scatter([t[1],t[2]],[f[1],f[2]], color='blue')

plt.subplot(2,2,4)
plt.title('Logic "NAnd" is Linearly Separable')
plt.scatter(t[1:], f[1:], color='blue')
plt.scatter(t[0],f[0], color='red')
plt.plot((1.1,.1), 'g--')

plt.subplots_adjust(right = 1.75, top=1.75  )
plt.show()
```

The perceptron algorithm has a limitation. A single perceptron cannot correctly classify all inputs if they are not linearly separable. Nonlinearly separable problems could not be solved until the invention of the multilayer perceptron and backpropagation. 

```python
# and, or, xor, nand (leaving out nand bc it is just the opposit of and)
# Learning And
and_inputs = [[1,1],[1,0],[0,1],[0,0]]
and_labels = [1,0,0,0]
and_perc = Perceptron(2,40)
and_perc.train(and_inputs, and_labels)
for x,y in zip(and_inputs,and_labels):
    print('input vector:\t', x, '\t\tprediction:\t', and_perc.predict(x), '\t\ttarget:\t', y)
```

```python
# Learning Or
or_inputs = [[1,1],[1,0],[0,1],[0,0]]
or_labels = [1,1,1,0]
or_perc = Perceptron(2,40)
or_perc.train(or_inputs, or_labels)
for x,y in zip(or_inputs, or_labels):
    print('input vector:\t', x, '\t\tprediction:\t', or_perc.predict(x), '\t\ttarget:\t', y)
```

```python
# Learning XOr
xor_inputs = [[1,1],[1,0],[0,1],[0,0]]
xor_labels = [0,1,1,0]
xor_perc = Perceptron(2,200)
xor_perc.train(xor_inputs, xor_labels)
for x,y in zip(xor_inputs, xor_labels):
    print('input vector:\t', x, '\t\tprediction:\t', xor_perc.predict(x), '\t\ttarget:\t', y)
```

The perceptron did not learn XOr despite the extra epochs because XOr is not linearly separable.
