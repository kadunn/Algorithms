# Solving the problem of linear separability and the XOr
By casting the data points into 3D space, the XOr problem can be learned by a single perceptron. Simply mutiply `x0 * x1` for each input:

```python
xor_inputs = [[1,1],[1,0],[0,1],[0,0]]
xor_labels = [0,1,1,0]
xor3D_inputs = [[xi[0],xi[1], xi[0]*xi[1]] for xi in xor_inputs]
print(xor3D_inputs)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter([1,0],[0,1],[0,0], color='red')
ax.scatter([1,0],[1,0],[1,0], color ='blue')

plt.show()
```

```python
xor3D_perc = Perceptron(3,30)
xor3D_perc.train(xor3D_inputs, xor_labels)
for x,y in zip(xor3D_inputs, xor_labels):
    print('input vector:\t', x, '\t\tprediction:\t', xor3D_perc.predict(x), '\t\ttarget:\t', y)
```

The perception has learned the XOr problem.
